```python
import numpy as np
import pandas as pd

import googlemaps
import gmplot
import folium

import matplotlib.pyplot as plt
```

```python
df_nodes = pd.read_table('new_nodes.txt', sep=' ', names=['index', 'names', 'lat', 'lng'])
df_nodes
```

```python
locations = folium.map.FeatureGroup()

for lat, lng, in zip(df_nodes['lat'], df_nodes['lng']):
    locations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=1,
            color='red',
        )
    )

# Add incidents to map
ucsd_map = folium.Map(location=[df_nodes['lat'][0], df_nodes['lng'][0]], zoom_start=50)
ucsd_map.add_child(locations)
```

```python
# selected_points = [1, 26, 51, 76, 101, 126, 151, 176, 201, 226]

selected_points = [226]
                   
locations = folium.map.FeatureGroup()

for i in selected_points:
    lat, lng = df_nodes.loc[i, 'lat'], df_nodes.loc[i,'lng']
    locations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=4,
            color='blue',
        )
    )

# Add incidents to map
ucsd_map = folium.Map(location=[df_nodes['lat'][i], df_nodes['lng'][i]], zoom_start=70)
ucsd_map.add_child(locations)
```
