```python
import pandas as pd
from skimage import io
import requests        #导入requests包
from bs4 import    BeautifulSoup
```

```python
df = pd.DataFrame(columns=['article', 'name', 'image', 'lat', 'lon', 'description', 'image_url'])
```

```python
f = open("NewYorkTourism.txt", 'r')
line00 = 1
while line00:
    line00 = f.readline()
    line01 = f.readline()
    line02 = f.readline()
    line03 = f.readline()
    line04 = f.readline()
    line05 = f.readline()
    line06 = f.readline()
    line07 = f.readline()
    line08 = f.readline()
    line09 = f.readline()
    line10 = f.readline()
    line11 = f.readline()
    line12 = f.readline()
    line13 = f.readline()
    pos = line01[5:-1]
    article = line02[9:-1]
    name = line03[6:-1]
    image = line04[7:-1]
    print(image)
    lat = line07[5:-1]
    lon = line08[5:-1]
    description = line10[14:-1].replace('[[', '').replace(']]', '')
    df = df.append({
        'article': article,
        'name': name,
        'image': image,
        'lat': lat,
        'lon': lon,
        'description': description,
    }
        , ignore_index=True)
f.close() 
```

```python
# df = df.loc[0:114]
```

```python
img_url = []
iteration = 0
```

```python
for img_name in df['image']:
    url = 'https://en.wikipedia.org/wiki/File:' + img_name
    strhtml = requests.get(url)
    soup = BeautifulSoup(strhtml.text,'lxml')
    data = soup.select('#mw-content-text > div.fullMedia > p > a')
    item = data[0]
    img_src = 'https:'+item.get('href')
    print(img_src)
    img_url.append(img_src)
    print('added to list')
    image = io.imread(img_src)
    io.imsave('./images/' + img_name, image)
    print('saved to local')
    iteration += 1
    print('progress: %d / 115' % iteration)
```

```python
df.loc[:,'image_url'] = img_url
```

```python
df.to_csv('NewYorkTourism.csv')
```
