```python
from cv2 import *
import urllib
import numpy as np
import requests as req
from PIL import Image
from io import BytesIO
from skimage import io
```

```python
url = 'https://en.wikipedia.org/wiki/File:AMEX.jpg'
strhtml = requests.get(url)
soup = BeautifulSoup(strhtml.text,'lxml')
data = soup.select('#file > div > span > a:nth-child(4)')
print(data)
```

```python
item = data[0]
```

```python
img_src = 'https:'+item.get('href')
```

```python
image = io.imread(img_src)
io.imshow(image)
io.show()
```

```python
io.imsave('./images/test.jpg', image)
```
