```python
import requests        #导入requests包
from bs4 import    BeautifulSoup

import pandas as pd

from cv2 import *
import urllib
import numpy as np
import requests as req
from PIL import Image
from io import BytesIO
from skimage import io
```

```python
df = pd.DataFrame(columns=['name', 'lat', 'lon', 'description', 'image_url'])
```

```python
url = 'https://en.wikipedia.org/wiki/Tourism_in_New_York_City'
strhtml = requests.get(url)
soup = BeautifulSoup(strhtml.text,'lxml')
    
for location in range(2,19):
    data = soup.select('#mw-content-text > div.mw-parser-output > table:nth-child(28) > tbody > tr:nth-child(' + str(location) + ') > td:nth-child(1) > a')
    if not data:
        continue
    item = data[0]
    name = item.get_text()
    print('name: ' + name)
    
    landmark_src = item.get('href')
    url = 'https://en.wikipedia.org' + landmark_src
    strhtml = requests.get(url)
    landmark_soup = BeautifulSoup(strhtml.text,'lxml')
    data = landmark_soup.select('#mw-content-text > div.mw-parser-output')
    item = data[0]
    geocode = item.find_all(name="span", attrs={"class" :"geo"})[0].get_text()
    geocode = geocode.split('; ')
    lat = geocode[0]
    lon = geocode[1]
    print('location: ' + str(lat) + ' ' + str(lon))
    
    data = landmark_soup.select('#mw-content-text > div.mw-parser-output')
    item = data[0]
    description_list = item.find_all(name="p")
    description = description_list[2].get_text()
    print('description: ' + description)

    data = soup.select('#mw-content-text > div.mw-parser-output > table:nth-child(28) > tbody > tr:nth-child(' + str(location) + ') > td:nth-child(2) > a > img')
    if not data:
        continue
    item = data[0]
    image = item.get('alt')
    url = 'https://en.wikipedia.org/wiki/File:' + image
    strhtml = requests.get(url)
    image_soup = BeautifulSoup(strhtml.text,'lxml')
    data = image_soup.select('#mw-content-text > div.fullMedia > p > a')
    if not data:
        continue
    item = data[0]
    img_src = 'https:'+item.get('href')
    print('img_src: ' + img_src)
    print(' ')
    
    df = df.append({
        'name': name,
        'image_url': img_src,
        'description': description,
        'lat': lat,
        'lon':lon
    }
        , ignore_index=True)
```
